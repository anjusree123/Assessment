```python
import cv2 
from cv2 import imread
from array import array
from numpy import *
from google.colab.patches import cv2_imshow
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/faces.JPG', cv2.IMREAD_UNCHANGED)
```

```python
detector_face = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
```

```python
if detector_face.empty():
    raise Exception("Failed to load cascade classifier.")
```

```python
imagem_cinza = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
face = detector_face.detectMultiScale(imagem_cinza, scaleFactor=1.3, minSize=(30,30))

for (x, y, w, h) in face:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 4)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,10))
plt.imshow(img_rgb)
plt.axis('off')
```
